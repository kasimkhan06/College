```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load Mall Customer Segmentation dataset
mall_data = pd.read_csv("Mall_Customers.csv")

# Select features for clustering
cluster_features = mall_data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize features
scaler = StandardScaler()
scaled_cluster_features = scaler.fit_transform(cluster_features)

# K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42 ,n_init=10)
mall_data['Cluster'] = kmeans.fit_predict(scaled_cluster_features)

# Visualize clusters
plt.scatter(mall_data['Annual Income (k$)'], mall_data['Spending Score (1-100)'], c=mall_data['Cluster'], cmap='viridis')
plt.title('K-Means Clustering on Mall Customer Data')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
